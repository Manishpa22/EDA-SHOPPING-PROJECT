```python
import pandas as pd   # use to clean data set
import numpy as np  # use in numerical
import matplotlib.pyplot as plt # use in visulise the data
import seaborn as sns # use visualise with advanced by matplolib
import plotly.express as px  # it also use same
import plotly.graph_objects as go # from plotly graph object create a dynamic graphy
import plotly.io as pio  # use creating template
import plotly.colors as cl
pio.templates.default = 'plotly_white'
```

# Data loading

# create column of date and month and year separate and also change datatype

```python
df['Order Date'] = pd.to_datetime(df['Order Date']) # remeber if u want to change the format or datatype use pd.to_date_time() OR use numpy
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
# to  create an column from extract use dt for date , time , month , year
df['Order Month'] = df['Order Date'].dt.month
df['Order Year'] = df['Order Date'].dt.year
df['Order Week'] = df['Order Date'].dt.dayofweek
```

```python
df.head()
```

```python
df.info()
```

```python
data = {
    1 : 'Jan', 2 : 'Feb', 3 : 'Mar', 4 : 'Apr', 5 : 'May', 6 : 'Jun', 7 : 'Jul', 8 : 'Aug',
    9 : 'Sep', 10 : 'Oct', 11 : 'Nov', 12 : 'Dec',
}
df['Order Month'] = df['Order Date'].dt.month
df['Order Month'] = df['Order Month'].map(data)
print(df['Order Month'])
```

```python
df.head()
```

# Monthly sales analysis


```python
sales_by_month = df.groupby(['Order Month'])['Sales'].sum().reset_index()
sns.lineplot(x='Order Month', y = 'Sales', data=sales_by_month)
plt.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=2, subplot_titles=["Seaborn-style Line", "Interactive Plotly"])

# First plot styled like Seaborn (static via Plotly)
fig.add_trace(go.Scatter(x=sales_by_month['Order Month'],   # scatter is use for lines plot
                         y=sales_by_month['Sales'],
                         mode='lines',
                         line=dict(color='black')),
              row=1, col=1)

# Second plot as Plotly default
fig.add_trace(go.Scatter(x=sales_by_month['Order Month'],
                         y=sales_by_month['Sales'],
                         mode='lines+markers',
                         line=dict(color='blue')),
              row=1, col=2)

fig.update_layout(title_text='Monthly Sales Analysis', height=600, width=1000)
fig.show()
```

***by above data we conclude that the minimun sales is in feb so that

# Sales by Category

```python
sales_by_category = df.groupby('Category')['Sales'].sum().reset_index()
print(sales_by_category)
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=1, cols=2, subplot_titles=["Sales by Category (Interactive)"],
                   specs=[[{'type':'domain'}, {'type':'domain'}]])  # in plotly plot if u want to make a pie chart then use specs

fg , ax =plt.subplots(1,1,figsize=(10,5))
plt.pie(sales_by_category['Sales'], labels=sales_by_category['Category'], autopct ='%1.1f%%', shadow=True, startangle=90, explode=[0,0.1,0])
plt.title('Sales by Category')
plt.show()

# second plot of plotly
fig.add_trace(go.Pie(labels=sales_by_category['Category'],   # if u want pie then use go.Pie and else use Scatter
                         values=sales_by_category['Sales'],
                         name='Sales by Category (Interactive)'),
              row =1, col =1)
fig.update_traces(textposition='inside',  hole =0.1, textinfo = 'percent+label')
fig.update_layout(title_text = 'Sales by Category', height=400, width=800)

fig.show()
```

*** In above analysis we conclude that technology have more sales compare to other

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px # Import plotly.express for potential future use if needed
from warnings import filterwarnings
filterwarnings('ignore')

sales_by_subcategory_sorted = sales_by_subcategory.sort_values(by='Sales', ascending=True)   # sort the data
# plot the graph here
fg , ax = plt.subplots(1,1,figsize=(20,7), facecolor='lightyellow')
sns.barplot(x='Sub-Category', y='Sales', data=sales_by_subcategory_sorted, palette='viridis', ax=ax)
for container in ax.containers:
  ax.bar_label(container, fmt='%.2f')
plt.title('Sales by Sub-category')
plt.tight_layout()
plt.show()
# Create a single subplot for the Plotly bar chart
fig = make_subplots(rows=1, cols=1, subplot_titles=["Sales by Sub-category (Interactive)"])

fig.add_trace(go.Bar(x=sales_by_subcategory['Sub-Category'],
                     y=sales_by_subcategory['Sales']),
              row=1, col=1)

# Update layout for title and dimensions
fig.update_layout(title_text='Sales by Sub-category', height=600, width=1000, plot_bgcolor='skyblue', paper_bgcolor='lightblue')  # for background or paper color use plot_bgcolor

fig.show()
```

# Monthly profit analysis

```python
df.head()
```

```python
profit_by_month = df.groupby('Order Month')[['Profit', 'Sales']].sum().reset_index()
print(profit_by_month)
```

```python
from warnings import filterwarnings
filterwarnings('ignore')
# sorted data of profit by month
profit_by_month_sorted = profit_by_month.sort_values(by='Profit', ascending=True)
# plot start here
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows =1, cols=2, subplot_titles=['Profit by month(line)', 'Profit by month(interactive)'])
#seaborn plot
fg , ax =plt.subplots(1,1,figsize=(10,5))
sns.barplot(x='Order Month', y='Profit', data=profit_by_month_sorted, hue='Sales', palette='viridis', edgecolor='black', ax=ax)
plt.title('Profit By Month(seaborn)', size=14)
plt.tight_layout()
for container in ax.containers:
  ax.bar_label(container, fmt='%.2f')
plt.show()

# first plot by graph
fig.add_trace(go.Scatter(x=profit_by_month['Order Month'],
                         y=profit_by_month['Profit'],
                         mode = 'lines+markers',
                         line=dict(color='black')),
              row=1, col=1)
# second plot by bar
fig.add_trace(go.Bar(x=profit_by_month_sorted['Order Month'],
                     y=profit_by_month_sorted['Profit'],
                     marker=dict(color='purple')), # Changed marker color to a single color
              row=1, col=2)
#update the style of grahy
fig.update_layout(title_text='Profit BY Month', height=600, width=1000, plot_bgcolor='lightgrey')
fig.show()
```

*** from above we conclusion that our profit by month is december and sales by month is November

# sales and profit by customer segment

```python
df.head()
```

```python
sales_profit_by_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit' : 'sum'}).reset_index()


color_palette = cl.qualitative.Vivid
fig = go.Figure()   # use for if u want to make a graph not separate
# first plot of sales by segment
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Sales'],
                     name= 'Sales',
                     marker_color = color_palette[0]))     # in that we not use rows, cols

# second plot of profit by segment
fig.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                     y=sales_profit_by_segment['Profit'],
                     name= 'Profit',
                     marker_color = color_palette[1]))

# update the layout and fig
fig.update_layout(title_text='Sales and Profit by Segment', paper_bgcolor='lightgrey', xaxis_title='Customer Segment', yaxis_title='Amount')
# plot show
fig.show()
```

*** from above we conclude that in customer segment the consumer is more purchase than other

# Sales to profit ratio

```python
sales_profit_by_segment = df.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
sales_profit_by_segment['sales_profit_ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'sales_profit_ratio']])
```
